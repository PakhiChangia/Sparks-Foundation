```python
#Importing all the libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
#Read the data
url=r"C:\Users\kc\Downloads\SampleSuperstore.csv"
df=pd.read_csv(url)
df
```

```python
# checking if there are any null values in the dataset
df.isnull().sum()
```

```python
#Checking the duplicate values
df.duplicated().sum()
```

```python
#Dropping the duplicate values
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.info()
```

```python
#Statitical information of the dataset
df.describe()
```

```python
df.drop(['Postal Code'],axis=1,inplace=True)
```

## Visualization

```python
sns.set_style("whitegrid")
sns.pairplot(df)
```

```python
print(df.corr())
sns.heatmap(df.corr(),annot=True,cmap="inferno")
plt.title("Correlation between Variables",size=20)
```

```python
plt.figure(figsize=(5,5))
plt.pie(df['Ship Mode'].value_counts(),labels=df['Ship Mode'].value_counts().index,autopct="%1.1f%%")
plt.title("Ship Mode",fontsize=25)
plt.show()
```

```python
plt.pie(df['Segment'].value_counts(),labels=df['Segment'].value_counts().index,autopct="%1.1f%%")
plt.title("Segment",size=20)
plt.show()
```

```python
plt.pie(df['Region'].value_counts(),labels=df['Region'].value_counts().index,autopct="%1.1f%%")
plt.title("Region",size=20)
plt.show()
```

```python
plt.pie(df['Category'].value_counts(),labels=df['Category'].value_counts().index,autopct="%1.1f%%")
plt.title("Category",size=20)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df['Sub-Category'].value_counts(),labels=df['Sub-Category'].value_counts().index,autopct="%0.1f%%")
plt.title("Sub-Category",size=20)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Sales with different Shipping Modes and Segments",size=15)
sns.lineplot(x="Ship Mode",y="Sales",data=df,ax=axes[0])
sns.lineplot(x="Segment",y="Sales",data=df,ax=axes[1])
plt.show()
```

```python
df.groupby("Category")["Profit","Sales"].agg(["sum"]).plot.bar()
plt.title("Total Profits and Sales based on different Category")
plt.show()
```

```python
df.groupby("Sub-Category")["Profit","Sales"].agg(["sum"]).plot.bar()
plt.title("Total Profits and Sales based on different Sub-Category")
plt.show()
```

```python
df.groupby("Region")["Profit","Sales"].agg(["sum"]).plot.bar()
plt.title("Total Profits and Sales based on Regions")
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total Sales Vs Profit",size=18)
sns.barplot("Category","Sales",hue="Region",data=df,ax=axes[0]).set_title("Category Vs Sales based on different Regions")
sns.barplot("Category","Profit",hue="Region",data=df,ax=axes[1]).set_title("Category Vs Profit based on different Regions")
```

```python
fig,axes=plt.subplots(2,1,figsize=(14,7))
fig.suptitle("Total Sales Vs Profits",size=18)
sns.barplot("Sub-Category","Sales",hue="Region",data=df,ax=axes[0])
sns.barplot("Sub-Category","Profit",hue="Region",data=df,ax=axes[1])
plt.show()
```

```python
plt.title("Discounts Vs Profits")
sns.lineplot(x="Discount",y="Profit",data=df)
```
